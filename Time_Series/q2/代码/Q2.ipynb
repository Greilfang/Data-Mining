```python
def set_prev_d(df):
    count =0 
    pluno_indexs = df.index.levels[0] 
    date_indexs= df.index.levels[1]
    for p in pluno_indexs:
        count = count+1
        print("p:{}{}".format(p,count))
        subset = df.loc[p]
        if subset.shape[0] == 1:
            continue
        else:
            subdates = list(reversed(subset.index))
            for i in range(len(subdates)):
                now_date = pd.to_datetime(subdates[i])
                for j in range(i+1,len(subdates)):
                    prev_date = pd.to_datetime(subdates[j])
                    #print(prev_date)
                    interval_days = int((now_date - prev_date).days)
                    if 0<interval_days<=7:
                        col = 'prev_d'+ str(interval_days)
                        df.at[(p,subdates[i]),col]=df.at[(p,subdates[i]),col]+ df.at[(p,subdates[j]),'qty']
                        if df.at[(p,subdates[j]),'qty']!=0:
                            print("index:",p,subdates[j],df.at[(p,subdates[j]),'qty'])
                            print(df.at[(p,subdates[i]),col])
                            print('-'*30)

                    else:
                        break
    
```

```python
def set_bnd_d(df):
    df['bnd_d1']=df['prev_d1']
    df['bnd_d2']=df['prev_d2']   
    df['bnd_d3']=df['prev_d3']
    df['bnd_d4']=df['prev_d4'] 
    df['bnd_d5']=df['prev_d5']   
    df['bnd_d6']=df['prev_d6']
    df['bnd_d7']=df['prev_d7']  
```

```python
def set_pl_d(df,lv):
    all_properties = np.unique([pl[:lv+1] for pl in df['pluno']])
    for prop in all_properties:
        print("prop:",prop)
        subset = df.loc[df['pluno'].str.startswith(prop),:]
        sorted_subset = subset.sort_values(by=['purchase_date'],ascending=[False])
        plength=len(sorted_subset.index)
        for i in range(plength):
            now_date = pd.to_datetime(sorted_subset.iloc[i]['purchase_date'])
            now = sorted_subset.index[i]
            for j in range(i+1,plength):
                prev_date = pd.to_datetime(sorted_subset.iloc[j]['purchase_date'])
                interval_days = (now_date - prev_date).days
                if 0<interval_days<=7:
                    col = 'pl{}_d{}'.format(lv,interval_days)
                    prev = sorted_subset.index[j]
                    df.at[now,col] =df.at[now,col]+df.at[prev,'qty']
                    if df.at[prev,'qty']!=0:
                        print(now,col,df.at[now,col])
                else:
                    break
```

```python
def set_pluno_week(df,wk):
    min_day = 7*(wk-1)
    max_day = min_day+7
    all_plunos = np.unique(df['pluno'])
    colmax = 'pl_{}week_max'.format(wk)
    colavg = 'pl_{}week_avg'.format(wk)
    colmin = 'pl_{}week_min'.format(wk)
#     tt=0
    for pluno in all_plunos:
        print("pluno:",pluno)
        subset = df.loc[df['pluno']==pluno,:]
        sorted_subset = subset.sort_values(by=['purchase_date'],ascending=[False])

        plength=len(sorted_subset.index)
        for i in range(plength):
            now_date = pd.to_datetime(sorted_subset.iloc[i]['purchase_date'])
            #print("now_date:",now_date)
            now = sorted_subset.index[i]
            count = 0
            for j in range(i+min_day,plength):
                prev_date = pd.to_datetime(sorted_subset.iloc[j]['purchase_date'])
                #print("prev_date:",prev_date)
                interval_days = int((now_date - prev_date).days)
                #print("intrval_days:",min_day,interval_days,max_day)
                if interval_days>max_day:
                    break
                elif min_day<interval_days<=max_day:
                    count = count+1
                    prev = sorted_subset.index[j]
                    if df.at[prev,'qty']>df.at[now,colmax]:
                        df.at[now,colmax] = df.at[prev,'qty']
                        #print("max:",df.at[now,colmax])
                    elif df.at[prev,'qty']<df.at[now,colmin]:
                        df.at[now,colmin] =df.at[prev,'qty']
                        #print("min:",df.at[now,colmax])
                    df.at[now,colavg]=df.at[now,colavg]+df.at[prev,'qty']
                    #print("avg:",df.at[now,colmax])

            if not count == 0: 
                df.at[now,colavg] = df.at[now,colavg]/count
            #print(df.at[now,colavg])
```

```python
def set_bndno_week(df):
    df['bnd_2week_avg'] = df['pl_2week_avg']
    df['bnd_2week_max'] = df['pl_2week_max']
    df['bnd_2week_min'] = df['pl_2week_min']
    df['bnd_3week_avg'] = df['pl_3week_avg']
    df['bnd_3week_max'] = df['pl_3week_max']
    df['bnd_3week_min'] = df['pl_3week_min']
    df['bnd_4week_avg'] = df['pl_4week_avg']
    df['bnd_4week_max'] = df['pl_4week_max']
    df['bnd_4week_min'] = df['pl_4week_min']
```

```python
def set_smaller_pluno_week(df,wk,lv):
    min_day = 7*(wk-1)
    max_day = min_day+7
    all_plunos = np.unique([pl[:lv+1] for pl in df['pluno']])
    colmax = 'pl{}_{}week_max'.format(lv,wk)
    colavg = 'pl{}_{}week_avg'.format(lv,wk)
    colmin = 'pl{}_{}week_min'.format(lv,wk)
    for pluno in all_plunos:
        print(pluno)
        subset = df.loc[df['pluno'].str.startswith(pluno),:]
        sorted_subset = subset.sort_values(by=['purchase_date'],ascending=[False])
        
        for date in reversed(np.unique(sorted_subset['purchase_date'])):
            #print("date:",date)
            indexs = sorted_subset[sorted_subset["purchase_date"]==date].index
            start_date = pd.to_datetime(date)
            before_dates =[start_date-dateutil.relativedelta.relativedelta(days=x) for x in range(min_day,max_day)]
            before_dates = [datetime.datetime.strftime(x,'%Y-%m-%d') for x in before_dates]
            calculated_data = sorted_subset[sorted_subset["purchase_date"].isin(before_dates)]
            df.loc[indexs,colmax]=np.max(calculated_data['qty'])
            df.loc[indexs,colavg]=np.mean(calculated_data['qty'])
            df.loc[indexs,colmin]=np.min(calculated_data['qty'])
```

```python
def my_feature_engineering_pipeline(df):
    print('step 1')
    df = df.set_index(['pluno','purchase_date'])
    set_prev_d(df)
    df =df.reset_index()
    set_bnd_d(df)
    
    print('step 2')
    
    set_pl_d(df,1)
    set_pl_d(df,2)
    set_pl_d(df,3)
    set_pl_d(df,4)
    
    print('step 3')
    
    set_pluno_week(df,2)
    set_pluno_week(df,3)
    set_pluno_week(df,4)
    
    print('step 4')
    set_bndno_week(df)
    set_smaller_pluno_week(df,2,1)
    set_smaller_pluno_week(df,3,1)
    set_smaller_pluno_week(df,4,1)
    set_smaller_pluno_week(df,2,2)
    set_smaller_pluno_week(df,3,2)
    set_smaller_pluno_week(df,4,2)
    set_smaller_pluno_week(df,2,3)
    set_smaller_pluno_week(df,3,3)
    set_smaller_pluno_week(df,4,3)
    set_smaller_pluno_week(df,2,4)
    set_smaller_pluno_week(df,3,4)
    set_smaller_pluno_week(df,4,4)
```

```python
my_feature_engineering_pipeline(df)
```
